```python
import random
import numpy as np

def toss_coins():
    # 0: Tail, 1: Head
    map = {0: 'T', 1: 'H'}

    # Toss 3 coins
    X1 = random.choice([0, 1])
    X2 = random.choice([0, 1])
    X3 = random.choice([0, 1])

    X = X1 + X2 + X3
    print(f"{map[X1]} {map[X2]} {map[X3]}   :   X = {X}")

    return X
```

```python
def update_empirical_statistics(n, current_mean, current_var, X):
    # Update the mean
    new_mean = current_mean + (X - current_mean) / n

    # Update the variance
    new_var = current_var + (X - current_mean) * (X - new_mean)
    new_var = (n / (n + 1)) * current_var + (X - current_mean) * (X - new_mean) / n
    return new_mean, new_var
```

```python
def simulate_tosses(num_tosses):
    mean = 0
    var = 0

    for i in range(1, num_tosses+1):
        # Toss 3 coins
        X = toss_coins()

        # "Empirical Mean" and "Variance"
        mean, var = update_empirical_statistics(i, mean, var, X)
        print(f"Empirical Mean = {mean:.2f} | Empirical Variance = {var:.2f} \n")
```

```python
simulate_tosses(100)
```
