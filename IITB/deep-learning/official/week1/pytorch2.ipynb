# Matrix Multiplication
1)Treat everything except the last two dims as batch dimensions

2)Last two dims must follow matrix multiplication rules:
(…, m, n) @ (…, n, p) → result (…, m, p)

3)Batch dimensions must be broadcastable according to PyTorch broadcasting rules

4)Broadcasting rules for batch dimensions:Batch dims must be equal or 1.If not, → RuntimeErroR

```python
import torch
#DIMENSION = (A,B,C,D) OR (C,D)
# Create two 2D tensors
A = torch.tensor([[1, 2],
                  [3, 4]], dtype=torch.float32)

B = torch.tensor([[5, 6],
                  [7, 8]], dtype=torch.float32)
#d=torch.tensor([[[1,2],[3,4]],
               # [[4,5],[5,7]]], dtype=torch.float32)
#print(d.shape)
print(A.shape)   #(B,r,c) where B= BATCHES , r=row, c=column
print(B.shape)

# Matrix multiplication
#print(d)
C = torch.matmul(A, B)  # or A @ B
print( C)
```

```python
import torch
#DIMENSION = (A,B,C,D) OR (C,D)
# Create two 2D tensors
A = torch.tensor([[1, 2,4],
                  [3, 4,5]], dtype=torch.float32)

B = torch.tensor([[5, 6],
                  [7, 8],[4,5]], dtype=torch.float32)
#d=torch.tensor([[[1,2],[3,4]],
               # [[4,5],[5,7]]], dtype=torch.float32)
#print(d.shape)
print(A.shape)   #(B,r,c) where B= BATCHES , r=row, c=column
print(B.shape)

# Matrix multiplication
#print(d)
C = torch.matmul(A, B)  # or A @ B
print( C)
```

```python
# Vector shape: (2,)
v = torch.tensor([1, 2], dtype=torch.float32)
print(v.shape)

# Matrix × Vector
result = torch.matmul(A, v)  # mv = matrix-vector
print("Matrix × Vector =", result)
```

# Dimension reduction

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]], dtype=torch.float32)  # shape (2, 3)

# Sum over all elements → scalar
print(torch.sum(x))  # shape: ()

# Sum over rows (dim=0) → keeps columns
print(torch.sum(x, dim=0))  # shape: (3,)

# Sum over columns (dim=1) → keeps rows
print(torch.sum(x, dim=1))  # shape: (2,)
```

```python
y = torch.randn(2, 1, 3, 1)  # shape: (2, 1, 3, 1)
#a=torch.randn(2,1,1,1,1,1,1,1,1,3,2)
y_squeezed = torch.squeeze(y)  # removes all size-1 dims
print(y_squeezed.shape)  # (2, 3)

# Remove only a specific axis
print(torch.squeeze(y, dim=1).shape)  # (2, 3, 1)
```

```python
import torch
# Create sample tensor
x = torch.arange(12).reshape(2, 2, 3)
print("Original shape:", x.shape)
print(x, "\n")


# Flattening / reshaping
a=torch.flatten(x)
print(a)
print("torch.flatten all:", a.shape)  # (12,)
b=torch.flatten(x, start_dim=1)
print(b)
print("torch.flatten from dim=1:", torch.flatten(x, start_dim=1).shape)  # (2, 6)
print("reshape alternative:", x.reshape(1, 12).shape)  # (1, 12)


```

# tensor to numpy and numpy to tensor


```python
import torch
import numpy as np

# Create a PyTorch tensor
t = torch.tensor([[1, 2, 3],
                  [4, 5, 6]], dtype=torch.float32)


# Convert to NumPy
n = t.numpy()

print(type(n))       # <class 'numpy.ndarray'>
print(n)
```

```python
# Create a NumPy array
n = np.array([[1, 2, 3],
              [4, 5, 6]], dtype=np.float32)

# Convert to tensor
t = torch.from_numpy(n)   # Shares memory
t = torch.tensor(n)  # Does NOT share memory

print(type(t))           # <class 'torch.Tensor'>
print(t)
```

# cpu to gpu and gpu to cpu

```python
import torch
import torch.nn as nn

# Simple model
model = nn.Linear(10, 5)  # Initially on CPU
print("Initial device:", next(model.parameters()).device)   #“Get the first parameter of the model, and tell me which device it’s on.”

# ---- Move model to GPU (if available) ----
if torch.cuda.is_available():
    model = model.to("cuda")
print("After moving to GPU:", next(model.parameters()).device)

# ---- Create a random input on GPU and run forward pass ----
if torch.cuda.is_available():
    x = torch.randn(4, 10, device="cuda")  # input on GPU
else:
    x = torch.randn(4, 10)
output = model(x)
print("Output device:", output.device)

# ---- Move model back to CPU ----
model = model.to("cpu")
print("After moving back to CPU:", next(model.parameters()).device)

# ---- Input also must be on CPU when model is on CPU ----
x_cpu = x.cpu()
output_cpu = model(x_cpu)
print("Output device after moving back to CPU:", output_cpu.device)
```
